```python
from logistic_regression import *
import matplotlib.pyplot as plt
```

Firstly, the logistic regression model will fit a linear separable data. Generated by random points divided in classes by the function y = 3x + 4

```python
def affine_function(x):
    return 3*x + 4
```

The data will be composed by the coordinades of 200 points.

```python
num_points = 200
X = np.random.randn(2,num_points)*10
Y = np.zeros((1,num_points))

for i in range(num_points):
    if X[1][i] > affine_function(X[0][i]):
        Y[0][i] = 1
        
print(Y)
```

```python
plt.scatter(X[0], X[1], c = Y[0], alpha=1)
plt.title('Scatter plot of affine data')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
params, costs = LogReg(X, Y, 0.5, 100000, print_every = 10000, print_cost = True, load_parameters = False, loaded_parameters = {})
```

```python
predictions = predict_LogReg(X, params)
print(Y - predictions)
```

After enough iterations, the model fitted the data !

Now, let's try with the famous XOR example

```python
xor_X = np.array([[0,0,1,1], [0,1,0,1]])
xor_Y = np.array([[0,1,1,0]])
```

```python
params, costs = LogReg(xor_X, xor_Y, 0.5, 1000, print_every = 100, print_cost = True, load_parameters = False, loaded_parameters = {})
```

```python
predictions = predict_LogReg(xor_X, params)
print(xor_Y - predictions)
```

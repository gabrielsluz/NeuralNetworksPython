Maybe normalizing the input can help

```python
def normalize_matrix(X):
    X_norm = np.zeros(X.shape)
    m = X.shape[1]
    lines = X.shape[0]
    
    X_average = np.zeros((lines, 1))
    X_variance = np.zeros((lines, 1))
    
    for i in range(m):
        X_average += X[:,i].reshape(lines,1)
        X_variance += X[:,i].reshape(lines,1)**2
            
    X_average = X_average / m
    X_variance = X_variance / m
    
    X_norm = (X - X_average) / X_variance
    
    return X_norm  
```

```python
X_norm = normalize_matrix(X)
plt.scatter(X_norm[0], X_norm[1], c = Y[0], alpha=1)
plt.title('Scatter plot of normalized affine data')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from vanilla_neural_networks import *
import matplotlib.pyplot as plt
```

Firstly, repeating the tasks used in the logistic regression notebook.

```python
def affine_function(x):
    return 3*x + 4
```

```python
num_points = 200
X = np.random.randn(2,num_points)*10
Y = np.zeros((1,num_points))

for i in range(num_points):
    if X[1][i] > affine_function(X[0][i]):
        Y[0][i] = 1
        
#print(Y)
```

```python
plt.scatter(X[0], X[1], c = Y[0], alpha=1)
plt.title('Scatter plot of affine data')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Logistic Regression
n_x = X.shape[0]
num_hidden_layers = 1
layers = [n_x, 1]
```

```python
parameters, costs = model_NN(X, Y, layers, learning_rate = 0.5, num_iterations = 10000 , print_cost = True, print_every = 1000, load_parameters = False, loaded_parameters = {})
```

```python
print(parameters)
```

```python
gradient_checking(X, Y, num_hidden_layers, parameters)
```
